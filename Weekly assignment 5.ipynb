# Predict the variable label (male or female)

```python
import sklearn as sk
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

Inlezen van de dataset

```python
df = pd.read_csv('voice.csv')
df.head()
```

Cleanen van de dataset

```python
df.info()
```

Opvallend dat de label nog een object is; omzetten naar dummies om de test te kunnen doen

```python
df = pd.get_dummies(df, drop_first = True)
df.info()
```

Correlatie bepalen

```python
df.corr()
```

Bij de kolommen'sd', 'Q75', 'IQR', 'sp.ent' en 'sfm' is een sterke correlatie te vinden. Die gebruiken we voor het model en de anderen worden verwijderd.

```python
df = df[['label_male','sd', 'Q75', 'IQR', 'sp.ent', 'sfm']]
df.info()
```

```python
df.corr()
```

```python
sns.violinplot(x='label_male', y='sd', data=df) #bekijken van de standaarddeviatie
```

Standaarddeviatie bij vrouwen ligt gemiddeld lager dan bij mannen

```python
sns.boxplot(x='label_male', y='sfm', data=df)
```

De sfm van mannen liggen gemiddeld hoger dan die van vrouwen

```python
sns.violinplot(x='label_male', y='sp.ent', data=df)
```

Er is duidelijk te zien dat mannen gemiddeld een hogere sp.ent hebben dan de vrouwen.

# Het model bouwen

```python
from sklearn.preprocessing import normalize
X = df.loc[:, ~df.columns.isin(['label_male'])] 
y = df['label_male'] 
X = normalize(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier 

knn = KNeighborsClassifier() 
knn = knn.fit(X_train, y_train) 
```

# Model evaluation

```python
knn.score(X_test, y_test)
```

Een hoge score van 92% accuraat

```python
from sklearn.metrics import confusion_matrix
y_test_pred = knn.predict(X_test) #the predicted values
cm = confusion_matrix(y_test, y_test_pred) #creates a "confusion matrix"
cm
```

```python
conf_matrix = pd.DataFrame(cm, index=['Vrouw (actual)', 'Man (actual)'], columns = ['Vrouw (predicted)', 'Man (predicted)']) 
conf_matrix
```
